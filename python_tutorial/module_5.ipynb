```python
# 302 - First episode starts before child was born.
# tables are header and episodes
# DOB from header and DECOM from episodes

header = dfs['header']

# ssda903 102 - date of birth is not a valid date

# coercing errors returns NATs for invalid dates
# if I slice out all NaTs - I have all the invalid dates!
header['DOB_dt'] = pd.to_datetime(header['DOB'], format="%d/%m/%Y", errors='coerce')

# .isna() rewutnrs true where the value in a row in a column is an Na
invalid_dobs = header['DOB_dt'].isna()
error_rows = header[invalid_dobs]

error_rows
```

```python
# 302 - First episode starts before child was born.
# tables are header and episodes
# DOB from header and DECOM from episodes

episodes = dfs['episodes']


episodes['DECOM_dt'] = pd.to_datetime(episodes['DECOM'], format="%d/%m/%Y", errors='coerce')

episodes.isna().sum()



# sorts values
episodes = episodes.sort_values(['CHILD', 'DECOM_dt'], ascending=[True,False])

#episodes = episodes.sort_values(by=['CHILD', 'DECOM_dt'], ascending=[True,False])
#episodes


# drops duplicates
episodes.drop_duplicates('CHILD', keep='first', inplace=True)
#episodes



df = pd.merge(episodes, header, how='left', on='CHILD', suffixes=['_epi', '_hdr'])

#df_1 = episodes.merge(header, on='CHILD', how='left', suffixes=['_epi', '_hdr'])
episodes


error_rows = df[df['DECOM'] < df['DOB']]
error_rows

```
