```python
#key : value - key must be immutable so either a string or number

fav_colours = {"Will":"black",
               "Clare": "yellow",
               "Marcel" : {"Monday": ["blue", "yellow"],
                           "Row":"Grey"}}
fav_colours["Will"]
fav_colours["Marcel"]["Monday"][0]
fav_colours.keys()
list[fav_colours.keys()]
print(list[fav_colours.values()])
print(list[fav_colours.items()])
```

```python
rest = {"KFC":["chicken","apple"],
        "MCD":{"Monday":["burger","fish"]}}
print(rest.keys())
print(rest.values())
print(rest.items())
```

```python
fav_colours = {"Will":"black",
               "Clare": "yellow",
               "Marcel" : {"Monday": ["blue", "yellow"],
                           "Row":"Grey"}}
fav_colours["Andrew"] = "black"
print(fav_colours)
fav_colours["Will"] = ["purple","black"]
print("Will" in fav_colours)
```
